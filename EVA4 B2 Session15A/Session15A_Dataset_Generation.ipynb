```
FGPATH='/content/dataset/fg'
BGPATH='/content/dataset/bg'
MASKPATH='/content/dataset/fg_mask'
OUTIMG='/content/gdrive/My Drive/Colab Notebooks/EVA4/Session14/image'
OUTMASK='/content/gdrive/My Drive/Colab Notebooks/EVA4/Session14/mask'
SUPERIMPOSE_MIN = 0
SUPERIMPOSE_MAX = 100
```

```
from PIL import Image
from random import randint
import numpy as np
from IPython.display import display
import zipfile
```

```
#Get a black backgorund image
black_arr = np.zeros((220, 220, 3))
blackbg = Image.fromarray(black_arr, mode='RGB')
img_counter = 0
#display(blackbg)

for bg_index in range (1, 101):
  out_zip =zipfile.ZipFile('/content/gdrive/My Drive/Colab Notebooks/EVA4/Session14/fg_bg.zip', mode='a', compression=zipfile.ZIP_STORED)
  # Open the bg image
  bg = Image.open(f'{BGPATH}/bg_{bg_index}.jpg')
  print("===== Reading the BG IMG :", bg_index, " =======")

  for fg_index in range(1, 101):
    # Open the fg image & fg mask
    fg = Image.open(f'{FGPATH}/fg_{fg_index}.png')
    fgmask = Image.open(f'{MASKPATH}/fg_{fg_index}.jpg').convert("RGBA")    
    fgflip = fg.transpose(Image.FLIP_LEFT_RIGHT)
    fgflipmask = fgmask.transpose(Image.FLIP_LEFT_RIGHT)

    print("For BG IMG: ", bg_index, "Reading the FG IMG : ", fg_index)

    # For each fg image, 20 superimpositions
    for randposimg in range (1, 21):
      print("For BG IMG: ", bg_index, "Reading the FG IMG : ", fg_index, "For Round ",randposimg)
      # Generate random (x, y) co-ordinates for superimposition
      x = randint(SUPERIMPOSE_MIN, SUPERIMPOSE_MAX)
      y = randint(SUPERIMPOSE_MIN, SUPERIMPOSE_MAX)
      # Make a copy of fg, bg and mask
      bg1 = bg.copy()
      bg2 = bg.copy()
      # Make a copy of the background image
      blackbg1 = blackbg.copy()
      blackbg2 = blackbg.copy()
      # Superimpose the images
      bg1.paste(fg, (x, y), fg)
      bg2.paste(fgflip, (x, y), fgflip)
      blackbg1.paste(fgmask, (x, y), fgmask) 
      blackbg2.paste(fgflipmask, (x, y), fgflipmask) 

      # Save the images 
      img_counter = img_counter + 1
      #print("Saving the Image Number: ", img_counter)
      bg1.save('bg1_temp.jpg', optimize=True, quality=30)
      blackbg1.save('blackbg1_temp.jpg', optimize=True, quality=30)
      # save the files to .zip file
      out_zip.write('bg1_temp.jpg',f'image/fg_bg_{str(img_counter)}.jpg')
      out_zip.write('blackbg1_temp.jpg',f'mask/fg_bg_mask_{str(img_counter)}.jpg')

      img_counter = img_counter + 1
      #print("Saving the Image Number: ", img_counter)
      bg2.save('bg2_temp.jpg', optimize=True, quality=30)
      blackbg2.save('blackbg2_temp.jpg', optimize=True, quality=30)      
      
      out_zip.write('bg2_temp.jpg',f'image/fg_bg_{str(img_counter)}.jpg')
      out_zip.write('blackbg2_temp.jpg',f'mask/fg_bg_mask_{str(img_counter)}.jpg')
  
  out_zip.close()
print("Total Images: ", img_counter)
```
